```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_csv("student.csv")

# Top 5 rows
print("Top 5 rows:")
print(data.head())

# Bottom 5 rows
print("\nBottom 5 rows:")
print(data.tail())

# Dataset info
print("\nDataset Info:")
data.info()

# Descriptive statistics
print("\nDescriptive Statistics:")
print(data.describe())

# Features (X) and Label (Y)
X = data[['Math', 'Reading']].values
Y = data['Writing'].values
```

```python
# X → shape (n, d)
# W → shape (d,)
# Y → shape (n,)
print("X shape:", X.shape)
print("Y shape:", Y.shape)
```

```python
def train_test_split_manual(X, Y, test_size=0.2, random_state=42):
    np.random.seed(random_state)
    indices = np.random.permutation(len(X))
    test_len = int(len(X) * test_size)

    test_idx = indices[:test_len]
    train_idx = indices[test_len:]

    return X[train_idx], X[test_idx], Y[train_idx], Y[test_idx]

X_train, X_test, Y_train, Y_test = train_test_split_manual(X, Y)
```

```python
def cost_function(X, Y, W):
    n = len(Y)
    Y_pred = np.dot(X, W)
    cost = (1 / (2 * n)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

```python
X_test_case = np.array([[1, 2], [3, 4], [5, 6]])
Y_test_case = np.array([3, 7, 11])
W_test_case = np.array([1, 1])

cost = cost_function(X_test_case, Y_test_case, W_test_case)

if cost == 0:
    print("Proceed Further")
else:
    print("Something went wrong")

print("Cost function output:", cost)
```

```python
def gradient_descent(X, Y, W, alpha, iterations):
    cost_history = [0] * iterations
    m = len(Y)

    for i in range(iterations):
        # Step 1: Prediction
        Y_pred = np.dot(X, W)

        # Step 2: Loss
        loss = Y_pred - Y

        # Step 3: Gradient
        dw = (1 / m) * np.dot(X.T, loss)

        # Step 4: Update weights
        W = W - alpha * dw

        # Step 5: Cost
        cost_history[i] = cost_function(X, Y, W)

    return W, cost_history
```

```python
np.random.seed(0)

X_rand = np.random.rand(100, 3)
Y_rand = np.random.rand(100)
W_rand = np.random.rand(3)

alpha = 0.01
iterations = 1000

final_params, cost_history = gradient_descent(X_rand, Y_rand, W_rand, alpha, iterations)

print("Final Parameters:", final_params)
print("Cost History (first 10):", cost_history[:10])
```

```python
def rmse(Y, Y_pred):
    return np.sqrt(np.mean((Y - Y_pred) ** 2))
```

```python
def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)
```
